```python
from ucimlrepo import fetch_ucirepo
import pandas as pd

# 1. Cargar el dataset
car_evaluation = fetch_ucirepo(id=19)

# 2. Obtener los datos originales
df = car_evaluation.data.features.copy()
df["Aceptacion"] = car_evaluation.data.targets["class"]

# 3. Renombrar las columnas al espa침ol
df.columns = [
    "Precio_Compra",        # buying
    "Costo_Mantenimiento",  # maint
    "Num_Puertas",          # doors
    "Capacidad_Personas",   # persons
    "Tamano_Maletero",      # lug_boot
    "Seguridad",            # safety
    "Aceptacion"            # class
]

# 4. Reemplazar valores con traducciones
df['Aceptacion'] = df['Aceptacion'].replace({
    'unacc': 'No aceptable',
    'acc': 'Aceptable',
    'good': 'Buena',
    'vgood': 'Muy buena'
})

df['Seguridad'] = df['Seguridad'].replace({
    'low': 'Baja',
    'med': 'Media',
    'high': 'Alta'
})

df['Precio_Compra'] = df['Precio_Compra'].replace({
    'vhigh': 'Muy alto',
    'high': 'Alto',
    'med': 'Medio',
    'low': 'Bajo'
})

df['Costo_Mantenimiento'] = df['Costo_Mantenimiento'].replace({
    'vhigh': 'Muy alto',
    'high': 'Alto',
    'med': 'Medio',
    'low': 'Bajo'
})

df['Tamano_Maletero'] = df['Tamano_Maletero'].replace({
    'small': 'Peque침o',
    'med': 'Mediano',
    'big': 'Grande'
})

df['Num_Puertas'] = df['Num_Puertas'].replace({
    '5more': '5 o m치s'
})

df['Capacidad_Personas'] = df['Capacidad_Personas'].replace({
    'more': 'm치s'
})

# 5. Guardar en CSV limpio
df.to_csv("car_data_limpio.csv", index=False)
```
